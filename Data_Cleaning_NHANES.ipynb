```python
import pandas as pd
import numpy as np
```

```python
initial_df = pd.read_csv('/content/NHANES_Extracted.csv')
initial_df.head()
```

```python
initial_df.isnull().sum()
```

```python
# Map for Education
education_map = {
    1: 'Less than 9th grade',
    2: '9-11th grade (Includes 12th grade with no diploma)',
    3: 'High school graduate/GED or equivalent',
    4: 'Some college or AA degree',
    5: 'College graduate or above',
    7: 'Refused',
    9: "Don't know"
}

# Map for Race/Ethnicity
race_map ={
    1: 'Mexican American',
    2: 'Other Hispanic',
    3: 'Non-Hispanic White',
    4: 'Non-Hispanic Black',
    6: 'Non-Hispanic Asian',
    7: 'Other Race - Including Multi-Racial'
}

# Map for Marital Status
marital_map = {
    1: 'Married/Living with partner',
    2: 'Widowed/Divorced/Separated',
    3: 'Never married',
    77: 'Refused',
    99: "Don't know"
}

# Map for Gender
Gender_map = {
    1: 'Male',
    2: 'Female'
}

# Frequency map
frequency_map = {
    1: 'Every day',
    2: 'Some days',
    3: 'Not at all',
    7: 'Refused',
    9: "Don't know"
}

# Yes/No map
yes_no_map = {
    1: 'Yes',
    2: 'No',
    3: 'Borderline',
    7: 'Refused',
    9: "Don't know"
}

# age filling map
age_desc_map = {
    666: 0,
    777: np.nan,
    999: np.nan
}
```

```python
initial_df['Highest level of education attained'] = initial_df['Highest level of education attained'].replace(education_map)
initial_df['Race/Hispanic origin category'] = initial_df['Race/Hispanic origin category'].replace(race_map)
initial_df['Marital status'] = initial_df['Marital status'].replace(marital_map)
initial_df['Gender of the participant (1=Male, 2=Female)'] = initial_df['Gender of the participant (1=Male, 2=Female)'].replace(Gender_map)
initial_df['Ever smoked at least 100 cigarettes in life?'] = initial_df['Ever smoked at least 100 cigarettes in life?'].replace(yes_no_map)
initial_df['Ever told by doctor they had Diabetes?'] = initial_df['Ever told by doctor they had Diabetes?'].replace(yes_no_map)
initial_df['Drank alcohol in the past 12 months?'] = initial_df['Drank alcohol in the past 12 months?'].replace(yes_no_map)
initial_df['Currently taking diabetic Pills to lower blood sugar?'] = initial_df['Currently taking diabetic Pills to lower blood sugar?'].replace(yes_no_map)
initial_df['Currently Smoke? (Every day, some days, or not at all)'] = initial_df['Currently Smoke? (Every day, some days, or not at all)'].replace(frequency_map)
initial_df['Age when first diagnosed with diabetes'] = initial_df['Age when first diagnosed with diabetes'].replace(age_desc_map)
```

```python
rows_to_calculate = initial_df[
    (initial_df['Body Mass Index (BMI: kg/m^2)'].isnull()) &
    (initial_df['Weight (kg)'].notnull()) &
    (initial_df['Standing Height (cm)'].notnull())
]
print(rows_to_calculate)
```

```python
rows_to_fill = initial_df[
    (initial_df['Currently Smoke? (Every day, some days, or not at all)'].isnull()) &
    (initial_df['Ever smoked at least 100 cigarettes in life?'] == 'No')
]
print(len(rows_to_fill))
```

```python
indices_to_fill = rows_to_fill.index
initial_df.loc[indices_to_fill, 'Currently Smoke? (Every day, some days, or not at all)'] = 'Not at all'
```

```python
condition = (initial_df['Ever told by doctor they had Diabetes?'] == 'No')
indices_to_fill = initial_df[condition].index
initial_df.loc[indices_to_fill, 'Currently taking diabetic Pills to lower blood sugar?'] = 'No'
initial_df.loc[indices_to_fill, 'Age when first diagnosed with diabetes'] = 0
```

```python
condition = initial_df['Drank alcohol in the past 12 months?'].isnull()
target_col = 'Average number of drinks per day when drinking'
initial_df.loc[condition, target_col] = initial_df.loc[condition, target_col].fillna(0)
```

```python
initial_df['Drank alcohol in the past 12 months?']=initial_df['Drank alcohol in the past 12 months?'].fillna('Not provided')
```

```python
initial_df['Marital status'] = initial_df['Marital status'].fillna('Not provided')
```

```python
initial_df['Currently Smoke? (Every day, some days, or not at all)'] = initial_df['Currently Smoke? (Every day, some days, or not at all)'].fillna('Not provided')
```

```python
initial_df['Ever smoked at least 100 cigarettes in life?'] = initial_df['Ever smoked at least 100 cigarettes in life?'].fillna('Refused')
```

```python
for col in initial_df.columns:
    if (initial_df[col].nunique() <= 10):
        print(col,": ", initial_df[col].unique())
```

```python
col_bmi = 'Body Mass Index (BMI: kg/m^2)'
median_bmi = initial_df[col_bmi].median()
initial_df[col_bmi].fillna(median_bmi, inplace=True)
```

```python
col_sbp = 'Systolic Blood Pressure, 1st reading (mmHg)'
median_sbp = initial_df[col_sbp].median()
initial_df[col_sbp].fillna(median_sbp, inplace=True)
```

```python
col_dbp = 'Diastolic Blood Pressure, 1st reading (mmHg)'
median_dbp = initial_df[col_dbp].median()
initial_df[col_dbp].fillna(median_dbp, inplace=True)
```

```python
numeric_cols_to_fill = [
    'Ratio of family income to poverty',
    'Weight (kg)',
    'Standing Height (cm)',
    'Glycohemoglobin (HbA1c) value (%)'
]

for col in numeric_cols_to_fill:
    median_value = initial_df[col].median()
    initial_df[col].fillna(median_value, inplace=True)
```

```python
categorical_cols_to_fill = [
    'Highest level of education attained',
    'Ever told by doctor they had Diabetes?',
    'Currently taking diabetic Pills to lower blood sugar?'
]

for col in categorical_cols_to_fill:
    initial_df[col].fillna('Not provided', inplace=True)
```

```python
sbp = initial_df['Systolic Blood Pressure, 1st reading (mmHg)']
dbp = initial_df['Diastolic Blood Pressure, 1st reading (mmHg)']

aha_conditions = [
    (sbp > 180) | (dbp > 120),
    (sbp >= 140) | (dbp >= 90),
    (sbp >= 130) | (dbp >= 80),
    (sbp >= 120) & (dbp < 80),
    (sbp < 120) & (dbp < 80)
]
aha_choices = [
    'Hypertensive Crisis',
    'Hypertension Stage 2',
    'Hypertension Stage 1',
    'Elevated',
    'Normal'
]

initial_df['AHA_Category'] = np.select(aha_conditions, aha_choices, default=None)

esh_conditions = [
    (sbp >= 180) | (dbp >= 110),
    (sbp >= 160) | (dbp >= 100),
    (sbp >= 140) | (dbp >= 90),
    (sbp >= 130) | (dbp >= 85),
    (sbp >= 120) | (dbp >= 80),
    (sbp < 120) & (dbp < 80)
]
esh_choices = [
    'Grade 3 Hypertension',
    'Grade 2 Hypertension',
    'Grade 1 Hypertension',
    'High-Normal',
    'Normal',
    'Optimal'
]

initial_df['ESH_Category'] = np.select(esh_conditions, esh_choices, default=None)

print("--- AHA Categories ---")
print(initial_df['AHA_Category'].value_counts(dropna=False))
print("\n--- ESH Categories ---")
print(initial_df['ESH_Category'].value_counts(dropna=False))
```

```python
df = initial_df.drop(['Marital status','Average number of drinks per day when drinking','Age when first diagnosed with diabetes'], axis=1)
```

```python
hba1c = df['Glycohemoglobin (HbA1c) value (%)']

conditions = [
    (hba1c >= 6.5),
    (hba1c >= 5.7)
]

choices = [
    'Diabetes',
    'Prediabetes'
]

df['Diabetes_Status'] = np.select(conditions, choices, default='Normal')
```
